```python
#I know there's a way to read only the paragraphs using = find_all(p) but I couldn't make it work using script below. Tried many, many times.

import requests
from bs4 import BeautifulSoup
from collections import Counter
import pandas as pd

Website1 = requests.get("http://www.indeed.com/jobs?q=data+scientist&l=San+Francisco+Bay+Area%2C+CA")
soup = BeautifulSoup(Website1.content)
texts = soup.find_all(text=True)
a = Counter([x.lower() for y in texts for x in y.split()])
b = (a.most_common())
makeaframe = pd.DataFrame(b)
makeaframe.columns = ['Words', 'Frequency']
print(makeaframe)
```
